**Exercise 0 Solution**

**Given:**
- Poisson equation: $\frac{\partial^2 u}{\partial x^2} = 0$ with $u(0) = u(1) = 0$
- Model: $u(x) = \sum_{k=-K}^K a_k \phi_k(x)$
- Basis functions: $\phi_0(x) = \frac{1}{\sqrt{2\pi}}$, $\phi_{-k}(x) = \frac{1}{\sqrt{\pi}}\cos(kx)$, $\phi_k(x) = \frac{1}{\sqrt{\pi}}\sin(kx)$

**Step 1: Apply the differential operator to each basis function**

$$N[\phi_0] = \frac{\partial^2}{\partial x^2}\left(\frac{1}{\sqrt{2\pi}}\right) = 0$$

$$N[\phi_k] = \frac{\partial^2}{\partial x^2}\left(\frac{1}{\sqrt{\pi}}\sin(kx)\right) = -k^2 \cdot \frac{1}{\sqrt{\pi}}\sin(kx) = -k^2 \phi_k$$

$$N[\phi_{-k}] = \frac{\partial^2}{\partial x^2}\left(\frac{1}{\sqrt{\pi}}\cos(kx)\right) = -k^2 \cdot \frac{1}{\sqrt{\pi}}\cos(kx) = -k^2 \phi_{-k}$$

**Step 2: Construct matrix A**

Matrix A has entries $A_{ij} = \langle N[\phi_i], N[\phi_j] \rangle$.

The Fourier basis functions are orthogonal, meaning:

$$\langle \phi_i, \phi_j \rangle = \int_0^{2\pi} \phi_i(x) \phi_j(x) \, dx = \begin{cases} 1 & \text{if } i = j \\ 0 & \text{if } i \neq j \end{cases}$$

This follows from the trigonometric identity $\sin(kx)\sin(mx) = \frac{1}{2}[\cos((k-m)x) - \cos((k+m)x)]$ and the fact that $\int_0^{2\pi} \cos(nx) \, dx = 0$ for $n \neq 0$.

**Diagonal entries ($i = j$):**

$$A_{kk} = \langle N[\phi_k], N[\phi_k] \rangle = \langle -k^2\phi_k, -k^2\phi_k \rangle = k^4 \langle \phi_k, \phi_k \rangle = k^4$$

**Off-diagonal entries ($i \neq j$):**

$$A_{ij} = \langle -i^2\phi_i, -j^2\phi_j \rangle = i^2 j^2 \langle \phi_i, \phi_j \rangle = 0$$

**Step 3: Compute condition number**

Matrix A is diagonal with entries $1^4, 2^4, 3^4, \ldots, K^4$.

$$\kappa(A) = \frac{\lambda_{\max}}{\lambda_{\min}} = \frac{K^4}{1} = K^4$$

**Conclusion:** The condition number grows as $K^4$. Higher frequency basis functions make the optimization problem severely ill-conditioned.
